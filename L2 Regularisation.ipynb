```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
X = []
for i in range(-100,100):
    X.append(i)
y = []
X = np.array(X)
# plt.plot(X,t)
X_copy = X
y = 5*X-(2*(X**2))+(3*(X**3))-X**4
X = X/100
# X_train,X_test,y_train,y_test
# X_min = X.min()
# X_max = X.max()
# X = (X - X_min) / (X_max - X_min)
d = 5
w = np.zeros(d)
```

```python
N = y.shape[0]
lda = 0.01
epochs = 1000
alpha = 0.1
# weight_change = np.zeros(w.shape[0])
for epoch in range(0,epochs):
    for sample in range(0,N):
        output = 0
        for feature in range(0,d):
            output+=w[feature]*pow(X[sample],feature)
        error = output-y[sample]
#         if error<0.1 and error>-0.1:
#             break
        for feature in range(0,d):
            sign = 0
            if(w[feature]>0):
                sign = 1
            elif(w[feature]<0):
                sign = -1
            w[feature]-= alpha*(2/N) * ( error*pow(X[sample],feature)+lda*w[feature]*sign)
print(w)
```

```python
predictions = []
for x in X: 
    out = 0
    for feature in range(0,d):
        out+=pow(x,feature)*w[feature]
    predictions.append(out)
```
