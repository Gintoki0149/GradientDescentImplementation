```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = []
for i in range(0,800):
    X.append(i)
t = []
X = np.array(X)
X_min = X.min()
X_max = X.max()
X = (X - X_min) / (X_max - X_min)
t = 5*X-(2*X**2)+(3*X**3)-6*X**4
d = 5
# t = np.sin(X)
# t_min = t.min()
# t_max = t.max()
# t = (t - t_min) / (t_max - t_min)
plt.plot(X,t)
w = np.zeros(d)
```

```python
N = t.shape[0]
lda = 0.00001
epochs = 1000
alpha = 0.1
weight_change = np.zeros(w.shape[0])
for epoch in range(0,epochs):
    for sample in range(0,N):
        output = 0
        for feature in range(0,d):
            output+=w[feature]*pow(X[sample],feature)
        error = output-t[sample]
        for feature in range(0,d):
            sign = 0
            if(w[feature]>0):
                sign = 1
            elif(w[feature]<0):
                sign = -1
            w[feature]-=(alpha*((2/N)*error*pow(X[sample],feature) + (lda/N)*sign))
print(w)
```

```python
out_array = []
for x in X:
    out = 0
    for j in range(0,d):
        out+=w[j]*pow(x,j)
    out_array.append(out)
plt.plot(X,out_array)
plt.plot(X,t)
```
